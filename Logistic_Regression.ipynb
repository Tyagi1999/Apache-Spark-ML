```
from pyspark.sql import SparkSession
```

```
spark=SparkSession.builder.appName('LogisticRegression').getOrCreate()
```

```
data=spark.read.format('libsvm').load('/FileStore/tables/sample_libsvm_data.txt')
```

```
data.describe().show()
```

```
data.show()
```

```
from pyspark.ml.classification import LogisticRegression
```

```
my_log_reg_model=LogisticRegression()
```

```
fitted_logreg=my_log_reg_model.fit(data)
```

```
log_summary=fitted_logreg.summary
```

```
log_summary.predictions.show()
```

```
lr_train,lr_test=data.randomSplit([0.7,0.3])
```

```
final_model=LogisticRegression()
```

```
fit_final=final_model.fit(lr_train)
```

```
test_result=fit_final.evaluate(lr_test)
```

```
test_result.predictions.show()
```

```
from pyspark.ml.evaluation import BinaryClassificationEvaluator,MulticlassClassificationEvaluator
```

```
evaluator=BinaryClassificationEvaluator()
```

```
my_final_roc = evaluator.evaluate(test_result.predictions)
```
